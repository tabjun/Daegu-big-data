##### 공공 기관 파이차트 

```python
plt.figure(figsize=(17,15))

#define data
data = [753,271,201,205,159,198,100,56,804,239,213,200,265,251,314,310,86]
labels = ['서울특별시','부산광역시','대구광역시','인천광역시','광주광역시','대전광역시',
          '울산광역시','세종특별자치시','경기도','강원도','충청남도','충청북도','전라남도',
          '전라북도','경상남도','경상북도','제주특별자치도']

wedgeprops={'width': 0.9, 'edgecolor': 'w', 'linewidth': 3}

plt.pie(data, labels = labels, autopct='%.0f%%', wedgeprops = wedgeprops)
plt.title('지역별 공공기관 지점 수', fontsize=15)
plt.show()
```
